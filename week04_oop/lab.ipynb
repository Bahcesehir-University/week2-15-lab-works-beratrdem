## Part 2: Core Concepts

### What is a Class?

A **class** is a blueprint for creating objects. It bundles
*attributes* (data) and *methods* (functions) into one unit.

Think of it like a template for a security badge:
the template defines what fields exist (name, clearance level, photo)
but each badge is a separate *object* filled with specific data.

````python
# A very simple class
class SecurityBadge:
    pass

# Create two objects (instances)
badge1 = SecurityBadge()
badge2 = SecurityBadge()

print(type(badge1))
print(badge1 is badge2)  # They are different objects
````

### The `__init__` Method

`__init__` runs automatically when you create an object.
It sets up the initial state (attributes) of that object.

The first parameter is always `self` — it refers to the
object being created.

````python
class SecurityBadge:
    def __init__(self, owner, level):
        self.owner = owner
        self.clearance_level = level

badge = SecurityBadge("Alice", 3)
print(badge.owner)
print(badge.clearance_level)
````

### Adding Methods

Methods are functions defined inside a class. They always
take `self` as the first parameter so they can access the
object's attributes.

````python
class SecurityBadge:
    def __init__(self, owner, level):
        self.owner = owner
        self.clearance_level = level

    def can_access(self, required_level):
        if self.clearance_level >= required_level:
            return True
        return False

badge = SecurityBadge("Alice", 3)
print(badge.can_access(2))   # True — level 3 >= 2
print(badge.can_access(5))   # False — level 3 < 5
````

### Exercise 1 — Create a `UserAccount` Class

Create a class called `UserAccount` with:
- An `__init__` method that accepts `username` and `email`
  and stores them as attributes.
- A method called `display_info` that prints the username
  and email in a readable format.

Then create two `UserAccount` objects and call `display_info`
on each.

````python
class UserAccount:

    def __init__(self, username, email):
        self.username = username
        self.email = email

    def display_info(self):
        print("Username:", self.username)
        print("Email:", self.email)
        print()

user1 = UserAccount("Alice", "alice@school.edu")
user2 = UserAccount("Janet", "janet@school.edu")

user1.display_info()
user2.display_info()
````

### Exercise 2 — Add a Password Check

Expand your `UserAccount` class (or create a new version) so that:

- Add a method `check_password(attempt)` that returns `True`

### Exercise 3 — Password Strength Validator

Create a class called `PasswordValidator` with:
- An `__init__` that accepts a password string.
- A method `is_long_enough()` that returns `True` if the
  password has 8 or more characters.
- A method `has_digit()` that returns `True` if the password
  contains at least one digit (0-9).
- A method `is_strong()` that returns `True` only if BOTH
  `is_long_enough()` and `has_digit()` return `True`.

Test with at least three different passwords.

````python
class PasswordValidator:
    def __init__(self, password):
        self.password = password

    def is_long_enough(self):
        return len(self.password) >= 8

    def has_digit(self):
        for char in self.password:
            if char.isdigit():
                return True
        return False

    def is_strong(self):
        return self.is_long_enough() and self.has_digit()

pass1 = PasswordValidator("pear")
print("apple is strong:", pass1.is_strong())

pass2 = PasswordValidator("peartree")
print("peartree is strong:", pass2.is_strong())

pass3 = PasswordValidator("peartree1")
print("peartree1 is strong:", pass3.is_strong())
````

### Exercise 4 — Log Entry

Create a class called `LogEntry` with:
- Attributes: `timestamp` (a string), `source_ip`, `message`
- A method `to_string()` that returns a single formatted string
  combining all three attributes.

Example output: `"[2025-03-23 10:15:00] 192.168.1.5 - Login attempt failed"`

Create at least two log entry objects and print them.

````python
class LogEntry:
    def __init__(self, timestamp, source_ip, message):
        self.timestamp = timestamp
        self.source_ip = source_ip
        self.message = message

    def to_string(self):
        return f"[{self.timestamp}] {self.source_ip} - {self.message}"

log1 = LogEntry("2025-03-23 10:15:00", "192.168.1.5", "Login attempt failed")
log2 = LogEntry("2025-03-23 10:16:22", "192.168.1.12", "Connection established")

print(log1.to_string())
print(log2.to_string())
````

### Challenge — Simple Intrusion Log

Create a class called `IntrusionLog` that:
- Stores a list of log entries (start with an empty list).
- Has an `add_entry(ip, message)` method that appends a
  dictionary with keys `"ip"` and `"message"` to the list.
- Has a `get_entries_by_ip(ip)` method that returns a list
  of only the entries matching a given IP address.
- Has a `count_entries()` method that returns the total
  number of entries.

Create an IntrusionLog, add at least 5 entries (some sharing
the same IP), then demonstrate each method.

````python
class IntrusionLog:
    def __init__(self):
        self.entries = []

    def add_entry(self, ip, message):
        new_log = {"ip": ip, "message": message}
        self.entries.append(new_log)

    def get_entries_by_ip(self, search_ip):
        matches = []
        for entry in self.entries:
            if entry["ip"] == search_ip:
                matches.append(entry)
        return matches

    def count_entries(self):
        return len(self.entries)

my_log = IntrusionLog()

my_log.add_entry("1.1.1.1", "Failed login attempt")
my_log.add_entry("2.2.2.2", "Port scan detected")
my_log.add_entry("3.3.3.3", "SQL injection attempt")
my_log.add_entry("1.1.1.1", "Malware signature match")
my_log.add_entry("1.1.1.1", "Multiple failed logins")
my_log.add_entry("2.2.2.2", "Error")
print("Total logs recorded:", my_log.count_entries())
print()

target_ip = "1.1.1.1"
print(f"Searching for logs from {target_ip}:")

found_logs = my_log.get_entries_by_ip(target_ip)
for log in found_logs:
    print(log)
````
